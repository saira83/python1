# nested loop

```python
#loop within the loop is called nested loop
for row in range(1, 15):
    print(row, "a"*row)
```

```python
for row in range(1, 27):
    print(row)
    for c in range(65, 65+row):
        print(chr(c))
```

```python
for row in range(1, 27):
    print(row)
    for c in range(65, 65+row):
        print(chr(c), end=" ")
```

```python
for row in range(1, 27):
    print(row, end="\t") # it used for spacing
    for c in range(65, 65+row):  # c is for charactors
        print(chr(c), end=" ")
    print()# by default space n seprater is included
```
